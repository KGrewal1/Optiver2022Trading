```python
import numpy as np
from matplotlib import pyplot as plt
MAXLIMIT = 190
#kindof sigmaesque
def sigmoid(L):
    return(3*(1-abs(L)/4)*L/(1+abs(2*L)))
S1 = sigmoid(1)
def limtransform(lmt):
    L = lmt/(2*MAXLIMIT)
    return((2*MAXLIMIT)*(S1-sigmoid(1-L)))
#none larger than 1
def margintransform(m):
    return((sigmoid(m)/S1))
lmt = np.arange(0,401, 1)
fig, ax = plt.subplots()



# plot the data
ax.plot(lmt,limtransform(lmt))
ax.set_xlabel('Distance From Limit')
ax.set_ylabel('Weighted Distance') # LaTeX syntax is also recognised. Do not worry if you have not heard of LaTeX, it is not important here

# set title using set_title
ax.set_title('Weight Function for Limit Distance')
ax.set_ylim([0,350])
ax.set_xlim([0,400])
# print the plot to screen


# if you want to save your figure, you can use the command plt.savefig('name of file')
plt.savefig('lmt.png', dpi=800)

plt.show()
```
